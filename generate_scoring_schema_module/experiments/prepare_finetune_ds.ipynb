```python
import pandas as pd

finetune_data_folder = "../data/finetuning_data/"

jd_to_scoring_schema_finetune_df = pd.read_csv(finetune_data_folder + "jd_to_scoring_schema_finetune_ds.csv")
jd_to_scoring_schema_finetune_df.head()
```

```python
finetune_samples = jd_to_scoring_schema_finetune_df[["prompt_sent", "raw_response_text"]]
finetune_samples.head()
```

```python
finetune_samples["prompt_sent"][0]
```

```python
finetune_samples["text_input"] = finetune_samples["prompt_sent"].str[9:-2]
finetune_samples["text_input"][0]
```

```python
finetune_samples.drop(columns=["prompt_sent"], inplace=True)
```

```python
# Swap 2 columns of the finetune_samples dataframe
finetune_samples = finetune_samples[["text_input", "raw_response_text"]]
finetune_samples.head()
```

```python
# Rename the 2 columns into "prompt" and "response"
finetune_samples.rename(columns={"text_input": "prompt", "raw_response_text": "response"}, inplace=True)
finetune_samples.head()
```

```python
# Save the finetune_samples dataframe to a CSV file
finetune_samples.to_csv(finetune_data_folder + "finetune_samples.csv", index=False)
```
